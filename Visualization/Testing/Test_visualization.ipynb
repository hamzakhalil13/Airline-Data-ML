```python
Q1 = df['NONSTOP_MILES'].quantile(0.25)
Q1

# # # Calculate Q3
Q3 = df['NONSTOP_MILES'].quantile(0.75)
Q3
# # # Define the Inter Quartile Range (IQR)
IQR = Q3 - Q1
IQR
# # # Make select condition for the values that fall below the Q1 - 1.5*IQR
outliers_below = df['NONSTOP_MILES'] < (Q1 - 1.5 * IQR)

# # # Make select condition for the values that fall above the Q3 - 1.5*IQR
outliers_above = df['NONSTOP_MILES'] > (Q3 + 1.5 * IQR)
outliers_above
# # # Select the INVERSE of the selection
df = df[ ~(outliers_above | outliers_below) ]
df.shape
```

```python
Q1 = test_df['MARKET_FARE'].quantile(0.25)
Q1

# # # Calculate Q3
Q3 = test_df['MARKET_FARE'].quantile(0.75)
Q3
# # # Define the Inter Quartile Range (IQR)
IQR = Q3 - Q1
IQR
# # # Make select condition for the values that fall below the Q1 - 1.5*IQR
outliers_below = test_df['MARKET_FARE'] < (Q1 - 1.5 * IQR)

# # # Make select condition for the values that fall above the Q3 - 1.5*IQR
outliers_above = test_df['MARKET_FARE'] > (Q3 + 1.5 * IQR)
outliers_above
# # # Select the INVERSE of the selection
test_df = test_df[ ~(outliers_above | outliers_below) ]
test_df.shape
```
