```python
df = pd.get_dummies(df, columns=['Origin', 'Dest', 'AirlineCompany'], drop_first=True)
df
```

```python
# for col in df.columns:
#     print(col)
    
my_list = list(df.columns)
# my_list
```

```python
selected_features = ['Miles']

X = df[selected_features]

# Set y to be our target variable. 
y = df['PricePerTicket']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)

# Print the length and width of our testing data.
print(X_train.shape, X_test.shape)
```

```python
params = { 
    'n_estimators' : [5, 10, 50, 100],
    'criterion' : ['gini', 'entropy'],
    'max_depth': [5, 10, 100], 
    'min_samples_split': [2, 10, 100], }
```

```python
grid_search_cv =  GridSearchCV( 
    estimator = RandomForestClassifier(), 
    param_grid = params, 
    scoring = 'f1' )
```

```python

# Train our model using our training data.

grid_search_cv.fit(X_train, y_train)
```

```python
model = grid_search_cv.best_estimator_
print(model)
```

```python
y_pred = model.predict(X_test)
```

```python
# Calculate our accuracy
accuracy = accuracy_score(y_test, y_pred)

# Calculate our precision score
precision = precision_score(y_test, y_pred)

# Calculate our recall score
recall = recall_score(y_test, y_pred)

f1 = f1_score(y_test, y_pred)

# Print each of our scores to inspect performance.
print("Accuracy Score: %f" % accuracy)
print("Precision Score: %f" % precision)
print("Recall Score: %f" % recall)
print('F1 Score %f' % f1)
```
