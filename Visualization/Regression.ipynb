```python
condition_1 = df['MARKET_FARE'] < 100

df = df[condition_1]
df
```

```python
mean_price = df['MARKET_FARE'].mean()
mean_price
```

```python
df = df.drop(columns= ['ORIGIN_STATE_NM', 'DEST_STATE_NM'  ], axis=1)

df
```

```python
df = pd.get_dummies(df, columns=['TICKET_CARRIER'], drop_first=True)
df
```

```python
selected_features = list(df.columns)
selected_features.remove('MARKET_FARE')
selected_features
```

```python
X = df[selected_features]

y = df['MARKET_FARE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
model_2 = RandomForestRegressor(random_state=1)

model_2.fit(X_train, y_train)

y_pred = model_2.predict(X_test)
```

```python
r_squared = metrics.r2_score(y_test, y_pred)
print('R-Squared Score:', r_squared)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, f1_score, roc_auc_score 
accuracy = model_2.score(X_test, y_test)
print("Model Accuracy: %f" % accuracy)
```

```python
model_3 = RandomForestRegressor(n_estimators=500, max_depth=10, random_state=1)
model_3.fit(X_train, y_train)

y_pred = model_3.predict(X_test)
```

```python
r_squared = metrics.r2_score(y_test, y_pred)
print('R-Squared Score:', r_squared)
```
